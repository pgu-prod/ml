Выводим статистику по числовым показателям

[pandas.DataFrame.describe](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html)

* count - количество наблюдений, отличных от NA/нулевых.
* max - максимум значений в объекте.
* min - минимум значений в объекте.
* mean- среднее значение.
* std - стандартное отклонение наблюдений.
* select_dtypes - подмножество DataFrame, включающее/исключающее столбцы в зависимости от их типа d.

Процентиль — значение, которое заданная случайная величина не превышает с фиксированной вероятностью, заданной в процентах.
Для числовых данных индекс результата будет включать count, mean, std, min, maxа также нижний 50и верхний процентили. По умолчанию нижний процентиль равен 25, а верхний процентиль равен 75. Процентиль 50аналогичен медиане.

[pandas.DataFrame.groupby](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html)


as_index bool, по умолчанию True - Возвращает объект с групповыми метками в качестве индекса. Актуально только для ввода DataFrame.

```python
data.groupby(by=['Make'],as_index=False).mean(numeric_only=True)
```

```python
data.groupby(by=['Year'],as_index=True).mean(numeric_only=True)
```
